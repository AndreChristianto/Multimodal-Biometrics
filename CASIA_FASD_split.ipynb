### Libraries

```python
import cv2 as cv
import os
```

### File Paths

```python
train_source_path = '../../Dataset/CASIA-FASD/train_release'
test_source_path = '../../Dataset/CASIA-FASD/test_release'
train_destination_path = '../../Dataset/CASIA-FASD/train'
test_destination_path = '../../Dataset/CASIA-FASD/test'
```

```python
def check_folder(folder):    
    if not os.path.exists(folder):
        os.makedirs(folder)
    else:
        print("Folder exists")
```

```python
check_folder(train_destination_path)
check_folder(test_destination_path)
```

### File Split

```python
def split_frame(folder, destination):
    for videos in os.listdir(folder):
        # Individual Selection
        videos_path = os.path.join(folder, videos)
        save_path = os.path.join(destination, videos)
        check_folder(save_path)

        genuine_count=0
        spoof_count=0
        
        # print(f"1.{videos_path}")
        for video in os.listdir(videos_path):
            # Video Selection
            clip = os.path.join(videos_path, video)
            print(f"2. |{video}|")

            cam = cv.VideoCapture(clip)

            fps = cam.get(cv.CAP_PROP_FPS)/2
            # print('frames per second = ', fps)

            if not cam.isOpened():
                print("Error: Could not open video file.")
                exit()

            n = 0
            # For genuine data
            if video == "1.avi" or video == "2.avi" or video == "HR_1.avi" or video == "HR_4.avi":
                while True:
                    # ret to make sure value exists, frame stores the image
                    # print(n)
                    ret, frame = cam.read()
                    if not ret:
                        break
                    if (n) % fps == 0 and ret == True:
                        print(f"genuine: {genuine_count}")
                        cv.imwrite(f"{save_path}/{genuine_count}_genuine.jpg", frame)
                        print(f"{save_path}/{genuine_count}_genuine.jpg")
                        genuine_count = genuine_count + 1
                    n = n + 1
            else:
                while True:
                    # ret to make sure value exists, frame stores the image
                    # print(n)
                    ret, frame = cam.read()
                    if not ret:
                        break
                    if (n) % fps == 0 and ret == True:
                        print(f"spoof: {spoof_count}")
                        cv.imwrite(f"{save_path}/{spoof_count}_spoof.jpg", frame)
                        print(f"{save_path}/{spoof_count}_spoof.jpg")
                        spoof_count = spoof_count + 1
                    n = n + 1

            cam.release()
            cv.destroyAllWindows
            
```

```python
split_frame(train_source_path, train_destination_path)
```

```python
split_frame(test_source_path, test_destination_path)
```
